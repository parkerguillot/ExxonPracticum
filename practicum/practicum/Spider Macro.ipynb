```python
#Imports scrapy process and settings
from scrapy.crawler import CrawlerProcess
from scrapy.utils.project import get_project_settings

#Places process and settings into process variable
process = CrawlerProcess(get_project_settings())
```

```python
#Calls the spiders that will be used
#Can comment out the spiders not needed or add spiders in this kernal
process.crawl('ceh')
process.crawl('ourhealth')
process.crawl('nationalgeo')
process.crawl('scienceblogs')
process.crawl('rwjf')
```

```python
#Starts the crawling process
process.start()
#Will need to restart the entire 
#notebook after every use of 'process.start()'
#to avoid 'ReactorNotRestartable' error
```

```python
#Notifies when scraping is done
print('Done Scraping')
```
