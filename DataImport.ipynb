```python
### current packages required for utilizing this notebook ###

import requests
from bs4 import BeautifulSoup
from spacy.lang.en import English
from spacy.tokenizer import Tokenizer
import pandas as pd
import numpy as np 
```

```python
### creates an object linked to the given url
url = 'https://www.energyindepth.org/germanys-energy-transition-makes-room-for-american-lng/'
```

```python
### retirieves the given object (url) from the web
r = requests.get(url)
```

```python
### converts the url html into text
html_doc = r.text
```

```python
soup = BeautifulSoup(html_doc, 'html5lib')
```

```python
soup.prettify()
```

```python
test_paragraphs = soup.find_all('p')
```

```python
balling = test_paragraphs[1].get_text()
```

```python
print(balling)
```

```python
def sometext(page):
    
    textlist = ''
    
    for s in page:
        w = s.get_text()
        textlist += w
    return textlist
```

```python
# def mytext(page):
#     for s in page:
#         return s.get_text()
```

```python
testing123 = sometext(test_paragraphs)
```

```python
print(testing123)
```

```python
nlp = English()
```

```python
doc = nlp(testing123)
```

```python
print(doc.text)
```

```python
tokenizer = Tokenizer(nlp.vocab)
```

```python
tokens = tokenizer(testing123)
```

```python
print(tokens)
```
